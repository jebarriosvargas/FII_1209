# Campo Eléctrico debido a un anillo


```python
from pylab import *
import plotly.graph_objects as go
import plotly
print(plotly.__version__)

a = 4
t = linspace(0,1)
z = a*cos(2*pi*t)
y = a*sin(2*pi*t)
x = zeros_like(t)

eje_x = linspace(0,6)
eje_y = zeros_like(x)
eje_z = zeros_like(x)


DATA = [
go.Scatter3d(x=x,y=y,z=z,mode='lines',
             line=dict(color="red",width=20),showlegend=False),
go.Scatter3d(x=eje_x,y=eje_y,z=eje_z,mode='lines',
             line=dict(color="black",width=4),showlegend=False)
]
fig = go.Figure(data=DATA)
fig.update_layout(
    title=r'$\\alpha$',
    width=400, height=400,
    margin=dict(t=40, r=0, l=20, b=20)
)

camera = dict(
    up=dict(x=0, y=1, z=2),
    center=dict(x=0, y=0, z=0),
    eye=dict(x=0.0, y=-2.5, z=0.0)
)

fig.update_layout(scene_camera=camera)

fig.show()
```
